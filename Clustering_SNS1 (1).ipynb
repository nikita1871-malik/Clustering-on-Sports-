```python
#Handling missings - Method2
def Missing_imputation(x):
    x = x.fillna(x.median())
    return x

sns1=sns1.apply(lambda x: Missing_imputation(x))
```

```python
#Handling Outliers - Method2
def outlier_capping(x):

    x = x.clip(upper=x.quantile(0.99))
    x = x.clip(lower=x.quantile(0.01))
    return x

sns2=sns1.apply(lambda x: outlier_capping(x))
sns2.describe(percentiles=[.9, .95, .96, .97, .98, .99]).T
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

def myKmeans(data,k, gap, verbose=True, grp=True):
    wcss=[]
    cluster=[]
    for i in range(1,k+1, gap):
        cluster.append(i)
        km=KMeans(n_clusters=i)
        km_result=km.fit(data)
        intr=km_result.inertia_
        wcss.append(intr)
        if verbose:
            print(f"for cluster {i} ; WCSS = {round(intr, 2)}")
    if grp:
        plt.plot(cluster, wcss, label="within cluster sum of sqaure(wcss)")
        plt.xlabel("Clusters")
        plt.ylabel("Wcss")
        plt.legend()
        plt.show()
    return wcss, cluster
```

```python
myKmeans(data=sns2,k=10, gap=1, verbose=True)
```
